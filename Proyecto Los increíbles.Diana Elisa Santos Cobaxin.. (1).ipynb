## Periferia, excentricidad, distancia, radio, etc
Debido a las relaciones entre los personajes y sus poderes, el grafo no es conexo lo que significa que no hay un camino directo que conecte todos los personajes entre sí. Por lo tanto, para calcular métricas como la excentricidad, periferia, distancia y radio en el grafo, primero necesitamos identificar y analizar cada componente conexo por separado. Estas métricas solo son calculables en los componentes conexos donde todos los personajes están directamente o indirectamente conectados.

```python
componentes_conexos = list(nx.connected_components(G))

# Para cada componente conexo, calcular excentricidad, periferia, distancia y radio
for i, comp in enumerate(componentes_conexos):
    subgrafo = G.subgraph(comp)
    print(f"Componente {i+1}:")
    print(f"   Nodos: {subgrafo.nodes}")
    print(f"   Excentricidad: {nx.eccentricity(subgrafo)}")
    print(f"   Periferia: {nx.periphery(subgrafo)}")
    print(f"   Distancia: {nx.average_shortest_path_length(subgrafo)}")
    print(f"   Radio: {nx.radius(subgrafo)}")
    print()
```

k es la constante de ajuste de la fuerza entre los nodos. Controla la distancia entre los nodos en la visualización. Valores más pequeños de k generan una disposición más compacta, mientras que valores mayores dispersan más los nodos.

```python
pos = nx.spring_layout(G, k=0.3)
```
